```python
# Q2

import ast

# Class for complex number parts
class NumComplex():
    def __init__(self, re, im):
        self.re = re
        self.im = im
        
    def __str__(self):
        if self.re == 0 and self.im == 0:
            return f"0"
        elif self.re == 0 and self.im == 1:
            return f"i"
        elif self.im == 1:
            return f"{self.re} + i" 
        if self.re == 0:
            return f"{self.im}i"
        else:
            return f"{self.re} + {self.im}i"

# Follows a basic multiplication formula
def ComplexMult(x, y):
    z = NumComplex(1, 1)
    z.re = x.re * y.re - x.im * y.im
    z.im = x.re * y.im + x.im * y.re
    return z

a = ast.literal_eval(input('Input first complex number as tuple: '))
b = ast.literal_eval(input('Input second complex number as tuple: '))
        
x = NumComplex(1, 1)
y = NumComplex(1, 1)
x.re = a[0]
x.im = a[1]
y.re = b[0]
y.im = b[1]

z = ComplexMult(x, y)

print("\n", f"({x})", f"({y})", " = ", z, sep='') 
    
import matplotlib.pyplot as plt
import numpy as np

# Plots original inputs and their product
real = z.re
imag = z.im
plt.xlim((-3, 3))
plt.ylim((-3, 3))
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.plot([0, real], [0, imag], 'ro-', label='python')
plt.text(real + 0.1, imag + 0.1, z)
plt.plot([0, x.re], [0, x.im], 'go-', label='python')
plt.text(x.re + 0.1, x.im + 0.1, x)
plt.plot([0, y.re], [0, y.im], 'bo-', label='python')
plt.text(y.re + 0.1, y.im + 0.1, y)

# Draws axis lines for aesthetic purposes
plt.axvline(0, linestyle = '--')
plt.axhline(0, linestyle = '--')

plt.show()

print("After multiplying the first complex number by i, the point rotates by 90 degrees (pi/2 radians)")
print("\n(2 + i)(i) = 2i + i^2 = -1 + 2i")
```

```python
# Q6

import mido
import time
import ast
import random

outport = mido.open_output(mido.get_output_names()[1])

# Input from user
notes_list = ast.literal_eval(input('Input MIDI notes: '))
notes_list = list(map(int, notes_list))
tempo = input('Tempo: ')
mode = input('Mode: ')
iterations = int(input('Iterations: '))

# Determines 'mode' chosen
if mode == 'up':
    notes_list.sort()
elif mode == 'down':
    notes_list.sort(reverse=True)
elif mode == 'updown':
    notes_list.sort()
    temp_list = notes_list.copy()
    notes_list.sort(reverse=True)
    temp_list += notes_list[1:]
    notes_list = temp_list.copy()
elif mode == 'random':
    random.shuffle(notes_list)
else:
    print("invalid entry - playing up")
    notes_list.sort()
 
# Converts tempo input into duration in seconds
note_length = mido.bpm2tempo(int(tempo))
note_length /= 1000000

# Sends series of MIDI messages for number of iterations
count = 0
while(count < iterations):
    if(len(notes_list)>0):
        for in_note in notes_list:
            on_msg = mido.Message('note_on', channel = 0, note = in_note, velocity = 64, time = 0)
            outport.send(on_msg)
            time.sleep(note_length)
            off_msg = mido.Message('note_off', channel = 0, note = in_note, velocity = 64, time = 0)
            outport.send(off_msg)
            time.sleep(0.001)
            outport.panic()
        count += 1
outport.panic()
```

```python
# Q7

import numpy as np
import matplotlib.pyplot as plt


# simple ADSR filter 
def adsr(freq, dur, length, decay, sus, srate = 44100):
    intervals = int(dur * freq)
    len_a = np.maximum(int(intervals * length[0]), 1)
    len_d = np.maximum(int(intervals * length[1]), 1)
    len_s = np.maximum(int(intervals * length[2]), 1)
    len_r = np.maximum(int(intervals * length[3]), 1)
    
    decay_a = decay[0]
    decay_d = decay[1]
    decay_s = decay[2]
    decay_r = decay[3]
    
    a = 1 / np.array([(1 - decay_a) ** n for n in range(len_a)])
    a = a / np.nanmax(a)
    d = np.array([(1 - decay_d) ** n for n in range(len_d)])
    d = d * (1 - sus) + sus
    s = np.array([(1 - decay_s) ** n for n in range(len_s)])
    s = s * sus
    r = np.array([(1 - decay_r) ** n for n in range(len_r)])
    r = r * s[-1]
    
    mags = np.concatenate((a, d, s, r))
    smooth = np.array([0.1 * (1 - 0.1) ** n for n in range(5)])
    smooth = smooth / np.nansum(smooth)
    mags = np.convolve(mags, smooth, mode='same')
    
    mags = np.repeat(mags, int(srate * dur / intervals))
    tail = int(srate * dur - mags.shape[0])
    if tail > 0:
        mags = np.concatenate((mags, mags[-1] - mags[-1] / tail * np.arange(tail)))
    return mags

# Creates list of all pitches in #-tet tuning
def tet_notes(base, tet, dur, adsr):
    for i in range(1, tet + 1):
        pitch = sinusoid((base * (2 ** (i / tet))), dur) # (2 ** (i / tet)) represents the 2^(1/#) formula
        note_list.append(pitch * adsr)

# Shouldn't be needed but was being glitchy otherwise
def drone_build(base, tet, dur, adsr):
    for i in range(1, tet):
        pitch = sinusoid(base, dur)
        drone_list.append(pitch)
    drone_list.append(sinusoid(base, srate = 44100, dur = dur, amp = 1) * adsr)
        
base = 110
tet = 12 # Species number of pitches in octave
dur = 1.5
adsr_values = adsr(440, dur, length=[0.1, 0.20, 0.50, 0.20], decay=[0.01,0.02,0.005,0.1], sus=0.2)
note_list = [sinusoid(base, dur) * adsr_values]
drone_list = [sinusoid(base, dur)]

tet_notes(base, tet, dur, adsr_values)
drone_build(base, tet, dur, adsr_values)

note_data = np.hstack(note_list)
drone_data = np.hstack(drone_list)
data_out = np.vstack([drone_data, note_data])
data_out2 = np.sum(data_out, axis=0)
ipd.Audio(data_out2, rate=44100, autoplay=True)

# Personally I've had a decent amount of experience with different tuning systems so they didnt sound very
# weird to me, but it is always interesting to hear them, especially comparing with other tets back to back
```

```python
# Q8

import ast
import numpy as np
import math
import cmath
import matplotlib.pyplot as plt

# Converts between time input and complex number form. Assumes a radius of 1
def TimeToComplex(num):
    r = 1
    angle = 90 - (num * 6)
    cnum = cmath.rect(r, math.radians(angle))
    if np.abs(cnum.real) < 0.000001:
        cnum = complex(0, cnum.imag)
    if np.abs(cnum.imag) < 0.000001:
        cnum = complex(cnum.real, 0)
    return cnum

# Draws a clock purely for aesthetic purposes
def ClockPlot():
    plt.plot(0, 0, 'ko-', label='python')
    plt.text(0, 1.3, "12", ha='center', va='center')
    plt.text(0.65, 1.13, "1", ha='center', va='center')
    plt.text(1.13, 0.65, "2", ha='center', va='center')
    plt.text(1.3, 0, "3", ha='center', va='center')
    plt.text(1.13, -0.65, "4", ha='center', va='center')
    plt.text(0.65, -1.13, "5", ha='center', va='center')
    plt.text(0, -1.3, "6", ha='center', va='center')
    plt.text(-0.65, -1.13, "7", ha='center', va='center')
    plt.text(-1.13, -0.65, "8", ha='center', va='center')
    plt.text(-1.3, 0, "9", ha='center', va='center')
    plt.text(-1.13, 0.65, "10", ha='center', va='center')
    plt.text(-0.65, 1.13, "11", ha='center', va='center')
    
    theta = np.linspace(0, 2 * np.pi, 100)
    r = 1.5
    cx = r * np.cos(theta)
    cy = r * np.sin(theta)
    plt.plot(cx, cy, 'k')
    

tmin, tsec = [int(x) for x in input('input a time in minutes and seconds, seperated by ":" ').split(':')]

cmin = TimeToComplex(tmin)
csec = TimeToComplex(tsec)

plt.xlim((-2, 2))
plt.ylim((-2, 2))
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.plot([0, cmin.real], [0, cmin.imag], 'b-', label='python')
plt.text(cmin.real + 0.1, cmin.imag + 0.1, "m", ha='center', va='center')
plt.plot([0, csec.real], [0, csec.imag], 'k-', label='python')
plt.text(csec.real + 0.1, csec.imag + 0.1, "s", ha='center', va='center')

ClockPlot()

plt.gca().set_aspect('equal')
plt.show()

print("In this particular plot, multiplying each value by 0.9945 - 0.1045i moves a minute clockwise")
print("In this particular plot, multiplying each value by 0.9945 + 0.1045i moves a minute counterclockwise")
```
